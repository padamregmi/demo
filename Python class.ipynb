```
cars=['audi','bmw','toyata','rolles royce','honda']
for car in cars:
  if car=='bmw':
    print(car.upper())
  else:
    print(car.title())
```

```
cars=['audi','bmw','toyata','rolles royce','honda']
for car in cars:
  if car!='bmw':
    print(car.upper())
  else:
    print(car.title())
```

```
for car in cars:
  if car=='bmw':
    print(car.upper())
  elif car=='audi':
    print(car.lower())
  else:
    print(car.title())
```
